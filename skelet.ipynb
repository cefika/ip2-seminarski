```python
counts = {}

# Iterirajte kroz cpai, npai i pai i brojite koliko se puta pojavljuje 6. atribut u svakom od njih
for data in unutracpai:
    count = data[6]
    if count in counts:
        counts[count][0] += 1
    else:
        counts[count] = [1, 0, 0, 0, 0, 0, 0, 0, 0]

for data in unutranpai:
    count = data[6]
    if count in counts:
        counts[count][1] += 1
    else:
        counts[count] = [0, 1, 0, 0, 0, 0, 0, 0, 0]

for data in unutrapai:
    count = data[6]
    if count in counts:
        counts[count][2] += 1
    else:
        counts[count] = [0, 0, 1, 0, 0, 0, 0, 0, 0]

for data in vannpai:
    count = data[6]
    if count in counts:
        counts[count][3] += 1
    else:
        counts[count] = [0, 0, 0, 1, 0, 0, 0, 0, 0]

for data in vancpai:
    count = data[6]
    if count in counts:
        counts[count][4] += 1
    else:
        counts[count] = [0, 0, 0, 0, 1, 0, 0, 0, 0]

for data in vanpai:
    count = data[6]
    if count in counts:
        counts[count][5] += 1
    else:
        counts[count] = [0, 0, 0, 0, 0, 1, 0, 0, 0]

for data in preseknpai:
    count = data[6]
    if count in counts:
        counts[count][6] += 1
    else:
        counts[count] = [0, 0, 0, 0, 0, 0, 1, 0, 0]

for data in presekcpai:
    count = data[6]
    if count in counts:
        counts[count][7] += 1
    else:
        counts[count] = [0, 0, 0, 0, 0, 0, 0, 1, 0]

for data in presekpai:
    count = data[6]
    if count in counts:
        counts[count][8] += 1
    else:
        counts[count] = [0, 0, 0, 0, 0, 0, 0, 0, 1]
        
podaci={}
# Ispišite rezultate
for count, (unutra_cpai, unutra_npai, unutra_pai, vannpai_count, vancpai_count, vanpai_count, presek_npai, presek_cpai, presek_pai) in counts.items():
    sumacpai=unutra_cpai+vancpai_count+presek_cpai
    sumanpai=unutra_npai+vannpai_count+presek_npai
    sumapai=unutra_pai+vanpai_count+presek_pai
    suma=sumacpai+sumanpai+sumapai
    podaci[count] = {
        "Tip:":"NC_009837",
        "CPai": unutra_cpai / sumacpai,
        "NPai": unutra_npai / sumanpai,
        "Pai": unutra_pai / sumapai,
        "VANNpai": vannpai_count / sumanpai,
        "VANCpai": vancpai_count /sumacpai,
        "VANPai": vanpai_count /sumapai,
        "PRESEKNpai": presek_npai / sumanpai,
        "PRESEKCpai": presek_cpai / sumacpai,
        "PRESEKPai": presek_pai / sumapai,
        "SUMA": suma
    }
    print(f"Tip: {df_fasta['id'][0]}, Niska: {count}, CPai: {unutra_cpai/sumacpai}, NPai: {unutra_npai/sumanpai}, VANNpai: {vannpai_count/sumanpai}, VANCpai {vancpai_count/sumacpai}, PRESEKNpai: {presek_npai/sumanpai}, PRESEKCpai: {presek_cpai/sumacpai}, SUMA: {suma}, SUMAC: {sumacpai}, SUMAN: {sumanpai}, SUMAP: {sumapai}")
```

```python
sortirani_podaci = sorted(podaci.items(), key=lambda x: x[1]["PRESEKCpai"],reverse=True)

# Prikaz sortiranih podataka
for niska, podatak in sortirani_podaci:
    print(niska, podatak)
```
