```python
import pickle
```

```python
with open('podaci.pkl', 'rb') as f:
    podaci = pickle.load(f)
```

```python
with open('podaci2.pkl', 'rb') as f:
    podaci2 = pickle.load(f)
```

```python
spojeni_podaci = {}
for kljuc in set(podaci) | set(podaci2):
    suma = sum([d.get('SUMA', 0) for d in [podaci.get(kljuc, {}), podaci2.get(kljuc, {})]])

    spojeni_podaci[kljuc] = {
        k: ((podaci.get(kljuc, {}).get(k, 0) * podaci.get(kljuc, {}).get('SUMA', 0) +
             podaci2.get(kljuc, {}).get(k, 0) * podaci2.get(kljuc, {}).get('SUMA', 0)) / suma)
        if suma != 0 else 0
        for k in ['CPai', 'NPai', 'VANNpai', 'VANCpai', 'PRESEK']
    }
    spojeni_podaci[kljuc]['SUMA'] = suma
```

```python
sortirani_podaci = sorted(spojeni_podaci.items(), key=lambda x: x[1]["SUMA"], reverse=True)

# Prikaz sortiranih podataka
for niska, podatak in sortirani_podaci:
    print(niska, podatak)
```
