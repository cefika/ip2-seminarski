```python
for index_fasta, row_fasta in df_fasta.iterrows():
    for index_cpai, row_cpai in df_cpai.iterrows():
        if (row_fasta['end_left'] < row_cpai['start']) or (row_fasta['start_left'] > row_cpai['end']):
            vancpai.add(tuple(row_fasta))
        elif (row_fasta['start_left'] >= row_cpai['start']) and (row_fasta['end_left'] <= row_cpai['end']):
            unutracpai.add(tuple(row_fasta))
        else:
            presekcpai.add(tuple(row_fasta))
        
        if (row_fasta['end_right'] < row_cpai['start']) or (row_fasta['start_right'] > row_cpai['end']):
            vancpai.add(tuple(row_fasta))
        elif (row_fasta['start_right'] >= row_cpai['start']) and (row_fasta['end_right'] <= row_cpai['end']):
            unutracpai.add(tuple(row_fasta))
        else:
            presekcpai.add(tuple(row_fasta))
```

```python
counts = {}

# Iterirajte kroz cpai, npai i pai i brojite koliko se puta pojavljuje 6. atribut u svakom od njih
for data in unutracpai:
    count = data[5]
    if count in counts:
        counts[count][0] += 1
    else:
        counts[count] = [1, 0, 0]

for data in unutranpai:
    count = data[5]
    if count in counts:
        counts[count][1] += 1
    else:
        counts[count] = [0, 1, 0]

for data in vancpai:
    count = data[5]
    if count in counts:
        counts[count][2] += 1
    else:
        counts[count] = [0, 0, 1]

# Ispišite rezultate
for count, (cpai_count, npai_count, pai_count) in counts.items():
    print(f"Br: {count}, CPai: {cpai_count}, NPai: {npai_count}, VAN: {pai_count}")
```
