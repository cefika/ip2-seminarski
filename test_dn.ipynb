```python
counts = {}

# Iterirajte kroz cpai, npai i pai i brojite koliko se puta pojavljuje 6. atribut u svakom od njih
for data in unutracpai:
    count = data[6]
    if count in counts:
        counts[count][0] += 1
    else:
        counts[count] = [1, 0, 0, 0, 0]

for data in unutranpai:
    count = data[6]
    if count in counts:
        counts[count][1] += 1
    else:
        counts[count] = [0, 1, 0, 0, 0]

for data in vannpai:
    count = data[6]
    if count in counts:
        counts[count][2] += 1
    else:
        counts[count] = [0, 0, 1, 0, 0]

for data in vancpai:
    count = data[6]
    if count in counts:
        counts[count][3] += 1
    else:
        counts[count] = [0, 0, 0, 1, 0]

for data in preseknpai:
    count = data[6]
    if count in counts:
        counts[count][4] += 1
    else:
        counts[count] = [0, 0, 0, 0, 1]
podaci={}
# IspiÅ¡ite rezultate
for count, (cpai_count, npai_count, vannpai_count, vancpai_count, presek_npai) in counts.items():
    suma=cpai_count+npai_count+vannpai_count+presek_npai+vancpai_count
    podaci[count] = {
        "CPai": cpai_count / suma,
        "NPai": npai_count / suma,
        "VANNpai": vannpai_count / suma,
        "VANCpai": vancpai_count /suma,
        "PRESEK": presek_npai / suma,
        "SUMA": suma
    }
    print(f"Niska: {count}, CPai: {cpai_count/suma}, NPai: {npai_count/suma}, VANNpai: {vannpai_count/suma}, VANCpai {vancpai_count/suma}, PRESEK: {presek_npai/suma}, SUMA: {suma}")
```

```python
sortirani_podaci = sorted(podaci.items(), key=lambda x: x[1]["SUMA"])

# Prikaz sortiranih podataka
for niska, podatak in sortirani_podaci:
    print(niska, podatak)
```

```python
with open('../podacitest/podaci2.pkl', 'wb') as f:
    pickle.dump(podaci, f)
```
